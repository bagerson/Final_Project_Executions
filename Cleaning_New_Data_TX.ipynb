```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings("ignore")
```

```python
new_data = "./DPIC-post-2002.csv" 
```

```python
new_data_df = pd.read_csv(new_data) 
```

```python
new_data_df.head()
```

```python
#select only the four columns from file_one_df
## year, race, gender, state, county, crimes committed (check open question), method of execution
new_data_selection = new_data_df[['Execution Date','Race','Sex','Execution Method']]
new_data_selection.head().dropna()
```

```python
all_dates = new_data_selection['Execution Date']
all_split = [date.split('/') for date in all_dates]
year = [arr[2] for arr in all_split]
new_data_selection['Year']=year
```

```python
new_data_selection=new_data_selection[['Year','Race','Sex','Execution Method']]
new_data_selection.head()
```

```python
new_data_selection.to_csv("./new_data_text.csv")
```

```python
EM_decode_dict = {"0M": "DoNotKnow;NA",
"1": "Hanging",
"2": "Electrocution",
"3" : "Gas",
"4" :"Firing Squad",
"5": "Lethal Injection",
"6": "Pressing",
"8": "Break on Wheel",
"10": "Burned",
"11": "Hung in Chains",
"13" : "Bludgeoned",
"14" :  "Gibbetted",
"15" :  "Other"}
EM_keys = EM_decode_dict.keys()
EM_values = EM_decode_dict.values()
EM_encode_dict=dict(zip(EM_values,EM_keys))
EM_encode_dict
```

```python
encoded_df=new_data_selection
replaced_EM = encoded_df["Execution Method"].replace(EM_encode_dict.keys(),value=EM_encode_dict.values())
replaced_EM.unique()
```

```python
encoded_df["Execution Method"]=replaced_EM
encoded_df.head()
```

```python
race_decode_dict={
"0M": "DoNotKnow;NA", 
"1" : "White",
"2" : "Black",
"3" : "Native American",
"4" : "Asian-Pacific Il",
"5" : "Latino",
"6" : "Other"}
race_encode_dict=dict(zip(race_decode_dict.values(),race_decode_dict.keys()))
race_encode_dict
```

```python
encoded_df['Race']=encoded_df["Race"].replace(race_encode_dict.keys(),value=race_encode_dict.values())
```

```python
encoded_df.head()
```

```python
sex_decode_dict={'0M':'DoNotKnow;NA','1': 'Male',
'2': 'Female'}
sex_encode_dict=dict(zip(sex_decode_dict.values(),sex_decode_dict.keys()))
```

```python
encoded_df['Sex']=encoded_df['Sex'].replace(sex_encode_dict.keys(),value=sex_encode_dict.values())
encoded_df
```

```python
encoded_df.to_csv("./new_data_encoded.csv")
```
